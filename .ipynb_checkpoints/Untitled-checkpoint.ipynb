```python
import time
start = time.time()

from openpyxl import  load_workbook
from openpyxl.styles import PatternFill
#from google.colab import files
import pandas as pd
```

```python
# _ = files.upload()
```

```python
df = pd.read_excel('Electricity.xlsx')
df.sample(5)
```

```python
df.shape
```

```python
df[['Consumption', 'Production']].isnull().sum()
```

```python
moy_cons = df['Consumption'].mean()
moy_prod = df['Production'].mean()
```

```python
wb = load_workbook('Electricity.xlsx')
ws =wb.active
```

```python
def test_func(colonne, moy_colonne):
    if colonne > moy_colonne :
        return '00FF00'
    elif colonne == moy_colonne :
        return 'FFFF00'
    elif colonne < moy_colonne :
        return 'FF0000'
```

```python
ws.cell(row=1, column=ws.max_column + 1, value="Status")
status = ''
i = 0
for row in ws.iter_rows(min_row=2,max_row= ws.max_row, min_col=1, max_col = ws.max_column):
    DateTime, Consumption, Production, Nuclear, Wind, Hydroelectric, Oil_Gas, Coal, Solar, Biomass, Status  = row
    cell_value_cons = Consumption.value
    cell_value_prod = Production.value
    if cell_value_cons > cell_value_prod :
        satuts = 'high consumption'
    else : 
        satuts = 'high production'
    color_cons = test_func(cell_value_cons, moy_cons)  
    color_prod = test_func(cell_value_prod, moy_prod)
    ws.cell(row=Consumption.row,column=2).fill = PatternFill(start_color=color_cons, end_color=color_cons, fill_type='solid')
    ws.cell(row=Production.row,column=3).fill = PatternFill(start_color=color_prod, end_color=color_prod, fill_type='solid')
    ws.cell(row=Consumption.row, column=ws.max_column + 1, value=status)
    i+=1
    print(i)


```

```python
wb.save('Electricity_formatted.xlsx')
```

```python
print(f'{round(time.time() - start)} s')
```
