```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill, Font
from openpyxl.comments import Comment
from utilities import add_comment
```

```python
old_path =  'yazaki_data/BOM/BOM_ref.xlsx'
new_path = 'yazaki_data/BOM/BOM_new.xlsx'
```

```python
df_old = pd.read_excel(old_path)
df_new = pd.read_excel(new_path)
```

```python
df_old
```

```python
len(list(df_old.columns)) - len(list(df_new.columns))
```

```python
columns = list(df_old.columns)
```

```python
print(f'Shape of new data : {df_new.shape}')
print(f'Shape of ref data : {df_old.shape}')
```

```python
df_new['_id'] = range(2, len(df_new) + 2)
df_old['_id'] = range(2, len(df_old) + 2)
```

### Unchanged Rows

```python
Unchanged_Rows = pd.merge(df_old, df_new, on=columns[:-1], how = 'inner', indicator = True)
```

### Deleted Rows

```python
Deleted_Rows = pd.merge(df_old, df_new, on=columns[:-1], how ='left', indicator = True)
```

```python
Deleted_Rows = Deleted_Rows[Deleted_Rows['_merge'] == 'left_only'].drop(columns = '_merge')
```

```python
Deleted_Rows = Deleted_Rows.rename(columns = {'_id_x' : 'Row_old_data',
                                '_id_y': 'Row_new_data'})
Deleted_Rows
```

```python
value = 'Removed'
Deleted_Rows['Row_new_data'].fillna(value, inplace=True)
Deleted_Rows
```

### Added Rows

```python
Added_Rows = pd.merge(df_old, df_new, on=columns[:-1], how ='right', indicator = True)
```

```python
Added_Rows = Added_Rows[Added_Rows['_merge'] == 'right_only'].drop(columns = '_merge')
```

```python
Added_Rows = Added_Rows.rename(columns = {'_id_x' : 'Row_old_data',
                                '_id_y': 'Row_new_data'})
Added_Rows
```

```python
value = 'Added'
Added_Rows['Row_old_data'].fillna(value, inplace=True)
```

```python
Added_Rows['Row_new_data'].values
```

### Changed Rows

```python
Changed_Rows = df_old[~df_old.isin(pd.concat([Unchanged_Rows, Added_Rows, Deleted_Rows], ignore_index=True,axis=0))]
```

```python
Changed_Rows._id.values
```

#### Highlighting differences by coloring 
- Modified cells in ⚪ gray
- Added Rows in 🟢 green
- Deleted Rows in 🔴 red

##### 1. Added Rows in 🟢 green

```python
green_fill = PatternFill(start_color='00FF00', end_color='00FF00', fill_type='solid')
font = Font(color='FFFFFF')
```

```python
Added_Rows[['Row_old_data', 'Row_new_data']]
```

```python
wb_new = load_workbook(new_path)
ws_new = wb_new.active

idx_added_rows = Added_Rows['Row_new_data'].values
for idx_row in idx_added_rows :
    for col in range(1, ws_new.max_column + 1):
        cell = ws_new.cell(row=idx_row, column=col)
        cell.fill = green_fill
    ws_new.cell(row=idx_row, column=1).comment = Comment("Added Row", "Author")
wb_new.save(new_path)
print(f'Added rows have been highlighted in green in {new_path}')
```

##### 2. Deleted Rows in 🔴 red

```python
red_fill = PatternFill(start_color='FF0000', end_color='FF0000', fill_type='solid')
font = Font(color='FFFFFF')
```

```python
Deleted_Rows[['Row_old_data', 'Row_new_data']]
```

```python
wb_old = load_workbook(old_path)
ws_old = wb_old.active

idx_Deleted_rows = Deleted_Rows['Row_old_data'].values
for idx_row in idx_Deleted_rows :
    for col in range(1, ws_old.max_column + 1):
        cell = ws_old.cell(row=idx_row, column=col)
        cell.fill = red_fill
    ws_old.cell(row=idx_row, column=1).comment = Comment("Deleted Row", "Author")
wb_old.save(old_path)
print(f'Removed rows have been highlighted in red in {old_path}')
```
